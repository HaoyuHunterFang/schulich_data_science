```python
import pandas as pd
```

```python
df1 = pd.read_csv('/Users/houhiroshisakai/Desktop/Professor Yang Research/cesd_item_level.csv')
df2 = pd.read_csv('/Users/houhiroshisakai/Desktop/Professor Yang Research/demog.csv')
df3 = pd.read_csv('/Users/houhiroshisakai/Desktop/Professor Yang Research/status_updates_cesd_v2.csv', encoding='ISO-8859-1', on_bad_lines='skip')
```

```python
df3['depression_label'].unique()
```

```python
df_temp = pd.merge(df1, df2, on='userid', how='outer')
df = pd.merge(df_temp, df3, on='userid', how='outer')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['depression_label'].fillna(0, inplace=True)
for i in df.columns:
    if df[i].dtype in ['int64','float64']:
        df[i].fillna(df[i].mean(), inplace=True)

    elif df[i].dtype == 'object':
        df[i].fillna('Other', inplace=True)
```

```python
df
```

```python
df['date_added'] = df['date_added'].astype(str)
df['time_completed'] = df['time_completed'].astype(str)
df['birthday'] = df['birthday'].astype(str)

#Convert 'Other' to timeFrame
specific_date = '2023-10-26'
specific_date_and_time = '2023-10-26 00:00:00'
df.loc[df['date_added'] == "Other", 'date_added'] = specific_date_and_time
df.loc[df['time_completed'] == "Other", 'time_completed'] = specific_date_and_time
df.loc[df['birthday'] == "Other", 'birthday'] = specific_date

# Convert type of datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['time_completed'] = pd.to_datetime(df['time_completed'], errors='coerce')
df['birthday'] = pd.to_datetime(df['birthday'], errors='coerce')
```
